## Analysis of centers - which one sells the most on average (weekly basis) and in total?

```python
centers_avg_sales = pd.read_sql_query("""
WITH sales AS
(
SELECT
t.week, t.center_id, SUM(t.checkout_price) AS total_weekly_sales
FROM test AS t
GROUP BY center_id, t.week
ORDER BY center_id
)
SELECT s.center_id, AVG(s.total_weekly_sales) AS average_weekly_sales
FROM sales AS s
GROUP BY center_id
ORDER BY average_weekly_sales DESC
""", conn)

centers_avg_sales
```

```python
centers_total_sales = pd.read_sql_query("""
SELECT
t.center_id, SUM(t.checkout_price) AS total_sales
FROM test AS t
GROUP BY center_id
ORDER BY total_sales DESC
""", conn)

centers_total_sales
```

## Promotion Analysis - How much in sales does it influence?

```python
count_of_promotions = pd.read_sql_query("""
SELECT
t.emailer_for_promotion AS promotion_use, ROUND((COUNT(t.checkout_price)/32573)*100, 2) AS proportion_of_transactions
FROM test AS t
GROUP BY promotion_use
ORDER BY proportion_of_transactions
""", conn)

count_of_promotions.style.hide_index()
```

## Promotion Analysis - In which areas it's most used? for any particular cuisine or category?

```python
promotions_by_city = pd.read_sql_query("""
SELECT
c.city_code, COUNT(t.checkout_price) AS transactions_w_promotions
FROM test AS t
JOIN fulfilment_center_info AS c
oN c.center_id = t.center_id
WHERE t.emailer_for_promotion = '1'
GROUP BY city_code
ORDER BY transactions_w_promotions DESC
""", conn)

promotions_by_city
```

```python
promotions_by_region = pd.read_sql_query("""
SELECT
c.region_code, COUNT(t.checkout_price) AS transactions_w_promotions
FROM test AS t
JOIN fulfilment_center_info AS c
oN c.center_id = t.center_id
WHERE t.emailer_for_promotion = '1'
GROUP BY region_code
ORDER BY transactions_w_promotions DESC
""", conn)

promotions_by_region.style.hide_index()
```

```python
promotions_by_cuisine = pd.read_sql_query("""
SELECT
m.cuisine, COUNT(t.checkout_price) AS transactions_w_promotions
FROM test AS t
JOIN meal_info AS m
oN m.meal_id = t.meal_id
WHERE t.emailer_for_promotion = '1'
GROUP BY cuisine
ORDER BY transactions_w_promotions DESC
""", conn)

promotions_by_cuisine.style.hide_index()
```

```python
promotions_by_category = pd.read_sql_query("""
SELECT
m.category, COUNT(t.checkout_price) AS transactions_w_promotions
FROM test AS t
JOIN meal_info AS m
oN m.meal_id = t.meal_id
WHERE t.emailer_for_promotion = '1'
GROUP BY category
ORDER BY transactions_w_promotions DESC
""", conn)

promotions_by_category.style.hide_index()
```

```python
promotions_by_center = pd.read_sql_query("""
SELECT
c.center_id, COUNT(t.emailer_for_promotion) AS promotions_used
FROM fulfilment_center_info AS c
JOIN test AS t
ON c.center_id = t.center_id
WHERE t.emailer_for_promotion = 1
GROUP BY center_id
ORDER BY promotions_used DESC
""", conn)

promotions_by_center
```

## In which weeks are more promotions used? Do promotions drive more spending?

```python
promotions_by_week = pd.read_sql_query("""
SELECT
t.week, COUNT(t.checkout_price) AS number_of_transactions
FROM test AS t
WHERE t.emailer_for_promotion = 1
GROUP BY week
ORDER BY number_of_transactions DESC
""", conn)

promotions_by_week.style.hide_index()
```

```python
sales_avg_by_promotion = pd.read_sql_query("""
SELECT
t.emailer_for_promotion, t.week, AVG(t.checkout_price) AS weekly_avg_sales
FROM test AS t
GROUP BY emailer_for_promotion, week
ORDER BY week
""", conn)

sales_avg_by_promotion.style.hide_index()
```

# Analysis of cities (590,526) with best sales

## how useful are promotions in this areas?

```python
best_cities = pd.read_sql_query("""
SELECT 
c.city_code, t.week, AVG(t.checkout_price) AS avg_weekly_sales, t.emailer_for_promotion
FROM fulfilment_center_info AS c
JOIN test AS t
ON c.center_id = t.center_id
WHERE c.city_code = 590 OR c.city_code = 526
GROUP BY city_code, emailer_for_promotion, week
ORDER BY week, city_code
""", conn)

best_cities
```
